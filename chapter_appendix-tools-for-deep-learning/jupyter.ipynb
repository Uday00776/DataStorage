```python
arr=np.array([[10,2],[3,4]])
arr2=np.array([[5,6],[7,8]])
print(arr+arr2)
print(arr*arr2)
print(arr@arr2)
print(arr//arr2)
print(arr**arr2)
print(arr%arr2)
arr.transpose()
```

```python
x=np.array([[12,11,15],[21,25,50],[18,27,16]])
y=np.sort(x,axis=1) #row wise sorting
k=np.sort(x,axis=0) #column wise sorting
z=np.argsort(x,axis=1) #retreive indices in sorted manner relative to original matrix
print(z)
print(y)
```

```python
x=np.array([5,3,8,1])
y=np.sort(x)[::-1] #For sorting in descending order
z=np.argsort(x)
x.sort() #For sorting x array in place
print(y)
print(z)
```

```python
x=np.array([1,2,3,4,5])
print(np.max(x))
print(np.min(x))
print(np.sum(x))
print(np.mean(x))
print(np.median(x))
y=np.std(x)
print(f"{y:.2f}")
print(np.var(x))
print(np.prod(x))
```
